**question 12 (2pts)**

Create a Pandes Series called `mean_area_ordinal` that, for each row in `dataset`, contains the value `low` if the value for `mean area` is within [0,420[, the value `middle` if the value for `mean area` is within [420,550[ and the value `high` if the value for `mean area` is larger of equal to 550. Use the `.apply()` function.

````python
#start code here

def make_ordinal(x):
    if x["mean area"] < 420:
        return "low"
    elif x["mean area"] < 550:
        return "middle"
    else:
        return "high"
    
mean_area_ordinal = dataset.apply(make_ordinal,axis="columns")

#end code here

print(mean_area_ordinal)
````

**question 14 (2pts)**

To prepare for Scikit-learn model fitting we need to keep the feature vectors and the label separate. Create a variable `X` that contains all the 30 features in `dataset` and a variable `y` that contains the label.

````python
#start code here

X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]

#start code here

print(X.shape)
print(y.shape)
````

**question 15 (1pt)**

Use the Scikit-learn `StandardScaler` to standardize the feature values in `X` (assign the result to `X_norm`).

````python
from sklearn.preprocessing import StandardScaler

#start code here

X_norm = StandardScaler().fit_transform(X)

#start code here

print(X_norm)
````

**question 16 (1pt)**

What does the following code do? What is in `X_train`, `X_test`, `y_train` and `y_test`?

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state=0)
````

*Answer here*

It splits the data *at random* into a training set (feature vectors in X_train and label in y_train (80% of the data)) and a test set (feature vectors in X_test and label in y_test (20% of the data))

**question 17 (2pts)**

Use the training set to compute the opimal values for the modelparameters in a Logistic Regression model (with hyperparameter `random_state` set to 1, all other hyperparameter values should be set to their Scikit-learn default values).

````python
from sklearn.linear_model import LogisticRegression

#start code here

lr_model = LogisticRegression(random_state=1)
lr_model.fit(X_train,y_train)

#end code here
````

**question 18 (2pts)**

Given the fitted model, compute predictions on the test set and compute the Area Under the Receiver Operating Characteristic curve (AUC) for these predictions.

````python
from sklearn.metrics import roc_auc_score

#start code here

predictions = pd.DataFrame(lr_model.predict_proba(X_test),columns=["prob_0","prob_1"])

score_auc = roc_auc_score(y_test,predictions["prob_1"])

#end code here

print(score_auc)
````

**question 19 (2pts)**

Create a Pandas DataFrame `df_model_parameters` with a column `feature` that contains the feature names in `X` and a column `parameter_value` that contains the corresponding fitted modelparameter value.

````python
#start code here

df_model_parameters = pd.DataFrame()
df_model_parameters["feature"] = X.columns
df_model_parameters["parameter_value"] = lr_model.coef_[0]

#end code here

print(df_model_parameters)
````

**question 20 (1pt)**

Add a comlumn `parameter_value_abs` to DataFrame `df_model_parameters` that contains the absolute value of the modelparameter values in column `parameter_value`.

````python
#start code here

df_model_parameters["parameter_value_abs"] = df_model_parameters["parameter_value"].map(abs)

#end code here

print(df_model_parameters)
````

**question 21 (1pt)**
